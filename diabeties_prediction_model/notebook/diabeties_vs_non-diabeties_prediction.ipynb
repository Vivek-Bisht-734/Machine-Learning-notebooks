### Importing Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

### Data Collection and Analysis (PIMA Diabetes Dataset)

```python
# Loading the diabetes dataset to a pandas DataFrame
df = pd.read_csv("C:/Users/bisht/Downloads/diabetes.csv")
df.head()
```

```python
# Number of rows and cols in dataset
df.shape
```

```python
# Checking statistical measures of dataset
df.describe().round(2)
```

```python
# Checking null or missing values in dataset
df.isna().sum()
```

```python
# CHecking total no. of values for diabetic and non-diabetic
df['Outcome'].value_counts()
# 0 = Non - Diabetec
# 1 = Diabetec
```

```python
# Checking the feature avg value for both diabatic and non-diabetic
df.groupby('Outcome').mean()
```

```python
# Separating data and labels
X = df.drop(columns = 'Outcome', axis = 1)
y = df['Outcome']

# Standardizing Dataset
scaler = StandardScaler()
scaler.fit(X)
standardized_data = scaler.transform(X)
print(standardized_data)
```

```python
X = standardized_data
y = df.Outcome
print(X)
print(y)
```

```python
# Train Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, 
                                                    random_state = 2, stratify = y)
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Training Model
classifier = svm.SVC(kernel = 'linear')

# Training the Support Vector Machine Classifier
classifier.fit(X_train, y_train)
```

```python
# Model Evaluation

# Accuracy Score on training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, y_train)
print("Accuracy score of the training data : ", training_data_accuracy)
```

```python
# Accuracy Score on test data
X_test_prediction = classifier.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, y_test)
print("Accuracy score of the training data : ", testing_data_accuracy)
```

```python
# Making a Predictive System
input_data = (4,110,92,0,0,37.6,0.191,30)   # Non-Diabetic Data

# converting input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# Reshaping array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

# Standardizing the data
std_data = scaler.transform(input_data_reshaped)
```

```python
prediction = classifier.predict(std_data)
print(prediction)

if prediction == True:
    print('Diebetic')
else:
    print('Non-Diebetic')
```
